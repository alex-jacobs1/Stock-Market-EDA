# **Data Scrape**
##### Using List of all stocks within the stock market in order to download historical stock data from yahoo finance in order to create all encompassing stock data csv.




```python
stocks = pd.read_csv('/content/drive/MyDrive/stocks.csv')

stock_tickers = stocks["Ticker"].tolist()

stocksdf = pd.DataFrame()

for ticker in stock_tickers:  
    
    print( str(stock_tickers.index(ticker)) + str(' : ') + ticker, flush=True)  
    
    try:
        stock = []
        stock = yf.download(ticker, period='max')
        
        if len(stock) == 0:
            None
        else:
            stock['Ticker']= ticker
            stocksdf = stocksdf.append(stock,sort=False)
    except Exception:
        None
```

```python
stocksdf = stocksdf.reset_index()

data = pd.merge(stocksdf, stocks[['Ticker', 'Company', 'Sector', 'Country']], on='Ticker', how='left')

data.to_csv('/content/drive/MyDrive/stock_market_info.csv')
```
